## Description:

```python
data['Description'] = data['Description'].astype(str).str.strip().str.upper()
```

```python
data["Description"].value_counts().head(20)
```

```python
def categorize_violation(desc):

    if pd.isna(desc):
        return "UNKNOWN"

    desc = str(desc).upper()

    if "SPEED" in desc:
        return "SPEEDING"

    elif "REG." in desc or "REGISTRATION" in desc or "PLATE" in desc:
        return "REGISTRATION"

    elif "LIC." in desc or "LICENSE" in desc or "PERMIT" in desc:
        return "LICENSE"

    elif "STOP SIGN" in desc or "RED SIGNAL" in desc:
        return "TRAFFIC_CONTROL"

    elif "TELEPHONE" in desc:
        return "DISTRACTED_DRIVING"
    
    elif "TRAFFIC CONTROL DEVICE" in desc:
        return "TRAFFIC_CONTROL"

    elif "NEGLIGENT" in desc or "RECKLESS" in desc:
        return "RECKLESS_DRIVING"

    elif "LANE" in desc:
        return "LANE_VIOLATION"

    elif "STOP LIGHT" in desc or "RED TRAFFIC SIGNAL" in desc:
        return "TRAFFIC_SIGNAL"

    elif "LIGHT" in desc or "LAMP" in desc or "HEADLIGHT" in desc or "TAILLIGHT" in desc:
        return "VEHICLE_LIGHTING"

    elif "WINDOW TINT" in desc or "TINT" in desc:
        return "WINDOW_TINT"

    elif "ALCOHOL" in desc:
        return "DUI"

    elif "SEATBELT" in desc or "SEAT BELT" in desc:
        return "SEAT_BELT"

    elif "EQUIPMENT" in desc:
        return "VEHICLE_EQUIPMENT"

    elif "UNINSURED" in desc:
        return "INSURANCE"

    elif "PEDESTRIAN" in desc:
        return "PEDESTRIAN"

    elif "CHILD SAFETY SEAT" in desc:
        return "CHILD_SAFETY"

    elif "PARKING" in desc:
        return "PARKING"
    elif "YIELD" in desc:
        return "RIGHT_OF_WAY"
    elif "FOLLOWING" in desc:
        return "FOLLOWING_DISTANCE"
    elif "TURN" in desc:
        return "TURNING_VIOLATION"
    elif "ACCIDENT" in desc or "SCENE" in desc:
        return "ACCIDENT_RELATED"
    elif "TEXT" in desc or "EARPHONE" in desc:
        return "DISTRACTED_DRIVING"
    elif "SECURITY" in desc or "INSURANCE" in desc:
        return "INSURANCE"
    else:
        return "OTHER"
```

```python
data['Violation_category'] = data['Description'].apply(categorize_violation)
```

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

data['TimeOfDay'] = data['Hour'].apply(time_of_day)
```
